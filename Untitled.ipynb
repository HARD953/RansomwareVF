```python
# Importation des bibliothèques nécessaires
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Chargement du fichier Excel
file_path = 'data1.xlsx'  # Remplacez par le chemin de votre fichier
df = pd.read_excel(file_path)

# Encodage one-hot de la colonne `Target Audience`
df = pd.get_dummies(df, columns=['Target Audience'], drop_first=True)

# Vérification des valeurs manquantes et imputation si nécessaire
if df.isnull().sum().sum() > 0:
    df = df.fillna(0)  # Imputation simple, ajuster selon les données si nécessaire

# Séparation des caractéristiques (X) et de la cible ('Hacker')
X = df.drop(columns=['Hacker'])  # Exclure la colonne cible 'Hacker'
y = df['Hacker']  # La cible est la colonne 'Hacker'

# Calcul de la matrice de corrélation pour les caractéristiques (X)
correlation_matrix = X.corr()

# Visualisation de la matrice de corrélation avec une heatmap
plt.figure(figsize=(14, 12))
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Matrice de corrélation des caractéristiques')
plt.show()

# Identification des colonnes fortement corrélées
threshold = 0.8
corr_pairs = correlation_matrix.abs().unstack().sort_values(ascending=False)
high_corr_pairs = [(col1, col2) for (col1, col2), corr in corr_pairs.items() if corr > threshold and col1 != col2]

# Créer une liste des colonnes fortement corrélées à supprimer
to_drop = set()
for col1, col2 in high_corr_pairs:
    if col2 not in to_drop:
        to_drop.add(col2)

# Afficher les paires de colonnes fortement corrélées
print("Paires de colonnes fortement corrélées (corrélation > 0.8) :")
for pair in high_corr_pairs:
    print(pair)

# Suppression des colonnes identifiées
df_reduced = df.drop(columns=to_drop)

# Affichage des colonnes restantes après suppression
print("\nColonnes après suppression des fortement corrélées :")
print(df_reduced.columns)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix, roc_curve, roc_auc_score, ConfusionMatrixDisplay
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.preprocessing import LabelEncoder
import joblib  # Pour la sauvegarde du modèle

# Séparation des caractéristiques (X) et de la variable cible (y)
if 'Hacker' in df.columns:
    X = df.drop(columns=['Hacker'])
    y = df['Hacker']
else:
    print("Erreur : La colonne cible 'Hacker' n'est pas présente dans le DataFrame original.")

# Encodage de la cible si nécessaire
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)

# Encoder les variables catégorielles
X_encoded = pd.get_dummies(X)

# Normalisation des données (important pour KNN et SVM)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_encoded)

# Séparation des données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

# Initialiser les modèles
models = {
    'KNN': KNeighborsClassifier(),
    'Random Forest': RandomForestClassifier(),
    'Decision Tree': DecisionTreeClassifier(),
    'SVM': SVC(probability=True)  # probability=True pour obtenir les probabilités pour la courbe ROC
}

# Dictionnaire pour stocker les résultats
results = {
    'Model': [],
    'Accuracy': [],
    'ROC AUC': []
}

# Variables pour la sélection du meilleur modèle
best_model = None
best_accuracy = 0
best_auc = 0

# Fonction pour afficher les courbes ROC et la matrice de confusion
def plot_roc_curve(fpr, tpr, auc, model_name):
    plt.plot(fpr, tpr, label=f'{model_name} (AUC = {auc:.2f})')
    plt.plot([0, 1], [0, 1], linestyle='--', color='gray')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'Courbe ROC pour {model_name}')
    plt.legend(loc='lower right')

def plot_confusion_matrix(conf_matrix, model_name):
    disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix)
    disp.plot(cmap=plt.cm.Blues)
    plt.title(f'Matrice de confusion pour {model_name}')
    plt.show()

# Boucle pour entraîner et évaluer les modèles
for model_name, model in models.items():
    # Entraîner le modèle
    model.fit(X_train, y_train)
    
    # Prédictions
    y_pred = model.predict(X_test)
    
    # Mesures de performance
    accuracy = accuracy_score(y_test, y_pred)
    auc = roc_auc_score(y_test, model.predict_proba(X_test)[:, 1])
    
    # Enregistrer les résultats
    results['Model'].append(model_name)
    results['Accuracy'].append(accuracy)
    results['ROC AUC'].append(auc)
    
    # Afficher la matrice de confusion
    conf_matrix = confusion_matrix(y_test, y_pred)
    plot_confusion_matrix(conf_matrix, model_name)
    
    # Calculer et afficher la courbe ROC
    fpr, tpr, _ = roc_curve(y_test, model.predict_proba(X_test)[:, 1])
    plot_roc_curve(fpr, tpr, auc, model_name)
    
    # Sélection du meilleur modèle basé sur l'accuracy et l'auc
    if accuracy > best_accuracy and auc > best_auc:
        best_accuracy = accuracy
        best_auc = auc
        best_model = model

# Affichage des résultats de performance
results_df = pd.DataFrame(results)
print("Comparaison des modèles:")
print(results_df)

# Afficher la courbe ROC
plt.figure()
plt.title('Courbe ROC comparée')
for model_name in models.keys():
    plt.plot(fpr, tpr, label=model_name)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='lower right')
plt.show()

# Sauvegarde du meilleur modèle, scaler et label encoder
if best_model is not None:
    print(f"\nLe meilleur modèle est : {best_model.__class__.__name__}")
    joblib.dump(best_model, 'best_model.pkl')  # Sauvegarde du meilleur modèle
    joblib.dump(scaler, 'scaler.pkl')          # Sauvegarde du scaler
    joblib.dump(label_encoder, 'label_encoder.pkl')  # Sauvegarde du label encoder
    print("Le meilleur modèle, le scaler et le label encoder ont été sauvegardés.")
```
